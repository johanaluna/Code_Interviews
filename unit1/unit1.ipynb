```python
column_names=['name','landmass','zone','area','population','language','religion','bars','stripes','colours','red',
             'green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars',
             'crescent','triangle','icon','animate','text','topleft','botright']
len(column_names)
```

```python
df=pd.read_csv("flag.data", names=column_names)
df.shape
```

```python
df.head()
```

```python
#replace language number by names
df['language'] = df['language'].replace({1:'English', 2:'Spanish', 3:'French', 4:'German', 5:'Slavic', 6:'OtherIndo-European', 7:'Chinese', 8:'Arabic', 
               9:'Japanese/Turkish/Finnish/Magyar', 10:'Others'})
```

```python
df['religion'] = df['religion'].replace({0:'Catholic', 1:'OtherChristian', 2:'Muslim', 3:'Buddhist', 4:'Hindu',
               5:'Ethnic', 6:'Marxist', 7:'Others'})
```

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
# missing values in data are represented by ?
```

```python
import numpy as np
df = df.replace({"?": np.NaN})
```
